Nesse momento, podemos perceber que o DataFrame possui valores classificados como NaN

algumas operações podem ser afetadas por tais valores, como plot, manipulação de string...

Portanto, no processo de limpeza de dados, devemos sempre analisar se as demais informações dessas colunas são relevantes, ou se podemos simplesmente remover os dados por completo

```python
df.loc[df["user_description"].isna()]
```

```python
df.shape
```

o DataFrame possui no total 179108 linhas, se excluirmos todos os dados NaN, de ambas as colunas, estariamos excluindo um maximo de 47059 (desconsiderando que algumas linhas possam ter NaN em ambas colunas).

# Excluindo dados

```python
df_exclui = df.dropna(subset=["user_description", "user_location"])
```

```python
df_exclui.shape
```

```python
df_exclui.loc[df_exclui["user_description"].isna()]
```

```python
df_exclui.loc[df_exclui["user_location"].isna()]
```
